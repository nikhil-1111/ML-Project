```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv(r'mushroom_cleaned.csv')
```

```python
x = df.iloc[:,[0,1,2,3,4,5,6,7]].values
y = df.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
st = StandardScaler()
x_s_train = st.fit_transform(x_train)
x_s_test = st.fit_transform(x_test)
```

<h1>Logistic Regression</h1>

```python
log = LogisticRegression(solver='lbfgs', max_iter=5000, random_state=0)
```

```python
log.fit(x_train,y_train)
```

```python
y1_pred = log.predict(x_test)
```

```python
print(pd.DataFrame({'Actual': y_test, 'Predicted': y1_pred}).to_string())
```

```python
print(accuracy_score(y_test,y1_pred)*100)
```

<h1>Support Vector Machine(SVM)

```python
classi = SVC(kernel='rbf',random_state=0)
classi.fit(x_s_train,y_train)
```

```python
y2_pred = classi.predict(x_s_test)
```

```python
print(pd.DataFrame({'actual':y_test,'predicted':y2_pred}).to_string())
```

```python
print(accuracy_score(y_test,y2_pred)*100)
```

<h1>KNN


```python
kn = KNeighborsClassifier()
```

```python
classi1 = KNeighborsClassifier(n_neighbors=5 , metric='minkowski',p = 2)
```

```python
classi1.fit(x_s_train,y_train)
```

```python
y3_pred = classi1.predict(x_s_test)
```

```python
print(pd.DataFrame({'actual':y_test,'predicted':y3_pred}).to_string())
```

```python
print(accuracy_score(y_test,y3_pred)*100)
```

<h1>Naive Bayes Claddifier Algorithm

```python
gb = GaussianNB()
```

```python
gb.fit(x_s_train,y_train)
```

```python
y4_pred = gb.predict(x_s_test)
```

```python
print(pd.DataFrame({'actual':y_test,'predicted':y4_pred}).to_string())
```

```python
print(accuracy_score(y_test,y4_pred))
```

<h1>Decision Tree Classification Algorithm

```python
dt = DecisionTreeClassifier(criterion='entropy',random_state= 0 )
```

```python
dt.fit(x_s_train,y_train)
```

```python
y5_pred = dt.predict(x_s_test)
```

```python
print(accuracy_score(y_test,y5_pred))
```

```python
print(pd.DataFrame({'actual':y_test,'predicted':y5_pred}).to_string())
```

<h1>Random Forest Classification

```python
rt = RandomForestClassifier(n_estimators = 10 , criterion='entropy') 
```

```python
rt.fit(x_s_train,y_train)
```

```python
y6_pred = rt.predict(x_s_test)
```

```python
print(pd.DataFrame({'actual':y_test,'predicted':y6_pred}).to_string())
```

```python
print(accuracy_score(y_test,y6_pred)*100)
```
